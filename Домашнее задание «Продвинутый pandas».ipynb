## Задание 1

```python
log = pd.read_csv('visit_log.csv', sep = ';')
```

```python
def source_type_function(log):
    if log['traffic_source'] in ['yandex', 'google']:
        return 'organic'
    elif log['traffic_source'] in ['paid', 'email'] and log['region'] == 'Russia':
        return 'ad'
    elif log['traffic_source'] in ['paid', 'email'] and log['region'] != 'Russia':
        return 'other'
    else:
        return log['traffic_source']
```

```python
log['source_type '] = log.apply(source_type_function, axis = 1)
log.head(5)
```
